```python
from src import detect_faces, show_bboxes
from PIL import Image
```

```python
data = {
    'imgs':  ['./imgs/{}.JPG'.format(i) for i in range(1,11)],
    'bounding_boxes': [],
    'landmarks':[],
}
```

## Training

```python

for img in data['imgs']:

    print("Parsing "+img+"...")
    bb, lm = detect_faces(Image.open(img))
    data['bounding_boxes'].append(bb)
    data['landmarks'].append(lm)
```

## Save results

```python
import pickle

with open('./data.dat','wb+') as f:
    pickle.dump(data, f)
```
